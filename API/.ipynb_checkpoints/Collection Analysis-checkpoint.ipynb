```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_json('doodles.json')
```

```python
df.head()
```

```python
df = df[['block_timestamp', 'transaction_hash', 'value', 'token_address', 'token_id', 'from_address', 'to_address']]
df
```

```python
plt.hist(df["token_id"], bins = 1000)
plt.ylim(0, 100)
```

```python
token_occur = df['token_id'].value_counts()
df_token = token_occur.to_frame().reset_index()
df_token.columns = ['token_id', 'occurence']
df_token
```

```python
df_token['occurence'].describe()
```

```python
df_token['occurence'].value_counts().sort_values()
```

```python
df[df['token_id'] == 9968]
```

```python
output = df.groupby('token_id').apply(lambda x: json.loads(x.to_json(orient='records'))).to_dict()
output
```

```python
with open('doodles_cleaned.json', 'w', encoding='utf-8') as f:
    json.dump(output, f, ensure_ascii=False, indent=4)
# json.dumps(output, './doodles_cleaned.json')
# './doodles_cleaned.json',
```

## Now begins an analysis on users involved in the collection

```python
user_occur = df['to_address'].value_counts()
df_user = user_occur.to_frame().reset_index()
df_user.columns = ['to_address', 'occurence']
df_user
```

```python
df_user['occurence'].describe()
```

```python
df_user['occurence'].value_counts().sort_values()
```

```python
output_user = df.groupby('from_address').apply(lambda x: json.loads(x.to_json(orient='records'))).to_dict()
output_user.pop('0x0000000000000000000000000000000000000000')
output_user
```

## We want to make sure no one is transferring or selling nft to itself!!

```python
df_argo = df[df['to_address'] != df['from_address']]
df_argo
```

```python
df_argo = df_argo[df_argo['from_address'] != '0x0000000000000000000000000000000000000000']
df_argo
```

```python
df_argo = df_argo.sort_values(['token_id'], ascending=False)
df_argo
```

```python
df_argo = df_argo[['from_address', 'to_address']]
df_argo
```

```python
df_edge = df_argo.groupby(['from_address', 'to_address']).size().to_frame().reset_index()
# .sort_values(['0'], ascending=False)
df_edge.columns = ['from_address', 'to_address', 'occurence']
df_edge = df_edge.sort_values(['occurence'], ascending=False)
df_edge
```

```python
df_edge['occurence'].describe()
```

```python
wallet_pair_occur = df_edge['occurence'].value_counts()
wallet_pair_occur
```

## So, less than 15% of all wallet pairs transfer more than once

```python
df_argo_edge = df_edge[['from_address', 'to_address']]
df_argo_edge.columns = ['source', 'target']
df_argo_edge
```

```python
df_argo_edge_limit = df_argo_edge.head(1000)
df_argo_edge_limit
```

```python
df_argo_edge_limit.to_csv('./doodles_argo_edge_1000', encoding='utf-8', index=False)
```
